```python
import pandas as pd
from datetime import datetime
from wordcloud import WordCloud
```

```python
df = pd.read_csv('../data/cleaned_data/cleaned_training_tweets.csv')
df.head()
```

Prep for Word Cloud

```python
word_cloud = df['text'].str.split(' ', expand=True)
word_cloud.head()
```

```python
word_cloud_lower = word_cloud.applymap(lambda x: x.lower() if isinstance(x, str) else x)
word_cloud_lower
```

```python
list = word_cloud_lower.columns.to_list()
list
```

```python
word_cloud_lower[0].value_counts()
```

```python
word_cloud_lower[1].value_counts()
```

```python
total = []
for i in list:
    count = word_cloud_lower[i].value_counts()
    total.append(count)
```

```python
total
```

```python
word_count = pd.DataFrame(total)
word_count
```

```python
new_list = word_count.columns.to_list()
new_list
```

```python
excluded_words = ['a', '', '-', 'an', 'and', 'are', 'as', 'at', 'be', 'but', 'by', 'for', 'if', 'in', 'into', 'is', 'it', 'no', 'not', 'of', 'on', 'or', 'such', 'that', 'the', 'their', 'then', 'there', 'these', 'they', 'this', 'to', 'was', 'will', 'with', 'i']
```

```python
filtered_list = [word for word in new_list if word not in excluded_words]
filtered_list
```

```python
value = []
for word in filtered_list:
    sum = word_count[word].sum()
    value.append(sum)
```

```python
value
```

```python
final_df = pd.DataFrame({'word': filtered_list, 'value': value})
final_df = final_df.sort_values(by='value', ascending=False)
```
